```python
# Loading  Necessary  library 

from selenium import webdriver
import pandas as pd
from bs4 import BeautifulSoup
from time import sleep
import warnings
warnings.filterwarnings('ignore')
import time
import string

driver=webdriver.Chrome(r"C:\Users\thunderbolt\sof\chromedriver.exe")

Job_Title = []
Company = []
Rating = []
Location = []
Short_Description = []
Days_Posted = []
Salary = []
Brief_Description = []
crawl_timestamp = []
crawl_date = []
```

```python
from datetime import date
```

```python
# Getting   all the data of Data Analyst   jobs in  gurgaon location.


for i in range(0,140,10):

    driver.get("https://in.indeed.com/jobs?q=data+analyst&l=Gurugram%2C+Haryana&start="+ str(i))

    time.sleep(5)

    all_jobs = driver.find_elements_by_class_name('result')

    for job in all_jobs:
    
        result_html= job.get_attribute('innerHTML')
        soup = BeautifulSoup(result_html,'html.parser')
        
        
        time.sleep(1)
        
        task1 = time.gmtime()
        crawl_timestamp.append(time.strftime("%Y-%m-%d %H:%M:%S", task1))
        
        
        today = date.today()    
        task2 = today.strftime("%d %B %Y")
        crawl_date.append(task2) 
        
        
    
    
        try:
            title = soup.find("a" ,class_="jobtitle").text.replace('\n',' ')
            Job_Title.append(title)
        except:
            Job_Title.append('---')
        
    
    
        try:
            company = soup.find("span" ,class_="company").text.replace('\n',' ')
            Company.append(company)
        except:
            Company.append('---')
        

        try:
            rating = soup.find("span" ,class_="ratingsContent").text.replace('\n',' ')
            Rating.append(rating)
        except:
            Rating.append('---')
        
        try:
            location = soup.find("span" ,class_="location accessible-contrast-color-location").text.replace('\n',' ')
            Location.append(location)
        except:
            Location.append('---')
        
        
        try:
            salary = soup.find("span" ,class_="salaryText").text.replace('\n',' ')
            Salary.append(salary)
        except:
            Salary.append('Not Mentioned')
        

        
        try:
            days_posted = soup.find("span" ,class_="date").text.replace('\n',' ')
            Days_Posted.append(days_posted)
        except:
            Days_Posted.append('---')
        
  
        
        sum_div = job.find_elements_by_class_name("summary")[0]
        sum_div.click()
        
        time.sleep(5)
        
        
        try:
            job_desc = driver.find_element_by_id("vjs-desc").text.replace('\n',' ')
            Brief_Description.append(job_desc)
        except:
            Brief_Description.append('----')
        
```

```python
# Getting length of all variables.
print(len(crawl_timestamp),len(crawl_date),len(crawl_date),len(Job_Title),len(Company),len(Rating),len(Location),len(Salary),len(Days_Posted),len(Brief_Description))
```

```python
# Getting Final Data
Data_Analyst =pd.DataFrame({ 'Job Title' : Job_Title ,'Company Name' : Company,'Location Of Job' : Location ,' Salary' :Salary , 'Rating' : Rating ,   'Days Posted' : Days_Posted ,'Crawl Timestamp': crawl_timestamp , ' Date Stamp' : crawl_date ,'Job Description' : Brief_Description  })
Data_Analyst
```

# Saving the file

```python
Data_Analyst.to_csv("Data_Analyst2.csv")
```
